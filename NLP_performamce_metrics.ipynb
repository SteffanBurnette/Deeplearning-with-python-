###Cosine Similarities
**Cosine similarity** ranges from 0 to 1, where 1 means the texts are identical.

* **Cosine similarity** is a good measure for text similarity because it calculates the **cosine **of the angle between two vectors, which represents the text documents in a **multi-dimensional space**. This approach focuses on the **orientation (direction)** rather than the **magnitude (length)** of the vectors, making it effective for comparing documents of different lengths.
* It ranges from 0 to 1, where **1 indicates identical texts** and **0 indicates completely dissimilar texts**.
 This makes it particularly useful for applications like information retrieval and text classification, where understanding the degree of similarity between documents is important.

 Calculating the cosine of the angle between two vectors tells you how similar the two vectors are in terms of their direction.

 When a vector is projected onto another vector or onto a subspace defined by a matrix, the resulting projection shares the same direction as the target vector or lies within the subspace. Specifically:

Projection onto a vector: The resulting projection vector lies along the direction of the target vector.
Projection onto a subspace: The resulting projection lies within the subspace defined by the matrix, aligning with one or more directions within that subspace.

 ### **Why cosine similarities over precision, recall and f1score**
*  **Cosine similarity** is more suitable for comparing two bodies of text because it **directly measures the similarity between the texts based on their content, ignoring differences in length and scale**. This is different from **precision, recall, and F1 score**, which are typically used for **classification tasks** where the focus is on the accuracy of predicting specific categories or labels. Cosine similarity, on the other hand, focuses on the overall similarity in the textual content, making it ideal for tasks like document comparison, information retrieval, and text clustering.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Combine original_data and generated_response into one DataFrame
text_data = pd.DataFrame({
    'text1': data['original_data'],
    'text2': data['generated_response']
})

# Compute TF-IDF vectors for both columns
vectorizer = TfidfVectorizer()
tfidf_matrix1 = vectorizer.fit_transform(text_data['text1'])
tfidf_matrix2 = vectorizer.transform(text_data['text2'])

# Compute cosine similarity between corresponding rows
cosine_similarities = []
for i in range(tfidf_matrix1.shape[0]):
    cosine_sim = cosine_similarity(tfidf_matrix1[i], tfidf_matrix2[i])
    cosine_similarities.append(cosine_sim[0][0])

# Add cosine similarity to the DataFrame
data['cosine_similarity'] = cosine_similarities

# Calculate the mean, median, and standard deviation of the cosine similarities
mean_similarity = data['cosine_similarity'].mean()
median_similarity = data['cosine_similarity'].median()
std_similarity = data['cosine_similarity'].std()

# Print the results
print(f"Mean Cosine Similarity: {mean_similarity:.2f}")
print(f"Median Cosine Similarity: {median_similarity:.2f}")
print(f"Standard Deviation of Cosine Similarity: {std_similarity:.2f}")

# Plot the distribution of cosine similarity
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(data['cosine_similarity'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Cosine Similarities')
plt.xlabel('Cosine Similarity')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### **BERTScore**
* Measures the similarity between two texts by calculating the similarity between their BERT embeddings.
* Strengths: accounts for semantic meaning, robust to noise.
* Weaknesses: computationally expensive, sensitive to embedding quality.
* Use case: compare text distributions, evaluate text generation.
* When to use: when you want a measure that accounts for semantic meaning and robustness, and you have access to BERT embeddings.

In general, you would choose a metric based on the specific requirements of your task:
If you want a simple, geometric measure of similarity, use Euclidean Distance.
If you want a set-based measure of similarity, use Jaccard Similarity.
If you want a widely accepted measure of similarity for machine translation and text summarization, use BLEU Score or ROUGE Score.

If you want a measure that accounts for semantic meaning and robustness, use METEOR Score, WMD, SPICE Score, or BERTScore.

```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util

# Load the BERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Combine original_data and generated_response into one DataFrame
text_data = pd.DataFrame({
    'text1': data['original_data'],
    'text2': data['generated_data']
})

# Compute BERT embeddings for both columns
embeddings1 = model.encode(text_data['text1'], convert_to_tensor=True)
embeddings2 = model.encode(text_data['text2'], convert_to_tensor=True)

# Compute cosine similarity between corresponding rows
bert_scores = util.cos_sim(embeddings1, embeddings2).diag()

# Add BERT scores to the DataFrame
data['bert_score'] = bert_scores.tolist()

# Calculate the mean, median, and standard deviation of the BERT scores
mean_bert = data['bert_score'].mean()
median_bert = data['bert_score'].median()
std_bert = data['bert_score'].std()

# Print the results
print(f"Mean BERT Score: {mean_bert:.2f}")
print(f"Median BERT Score: {median_bert:.2f}")
print(f"Standard Deviation of BERT Scores: {std_bert:.2f}")

# Plot the distribution of BERT scores
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(data['bert_score'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of BERT Scores')
plt.xlabel('BERT Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Create a list with 1 or 0 based on bert score
bert_labels = [1 if score > 0.6 else 0 for score in data['bert_score']]


actual_labels = [1] * len(data['bert_score'])

# Calculate precision, recall, and f1 score
precision = precision_score(actual_labels, bert_labels)
recall = recall_score(actual_labels, bert_labels)
f1 = f1_score(actual_labels, bert_labels)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(actual_labels, bert_labels), annot=True, cmap='Blues')
plt.xlabel('Predicted Labels')
plt.ylabel('Actual Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Combine original_data and generated_response into one DataFrame
text_data = pd.DataFrame({
    'text1': data['original_data'],
    'text2': data['generated_data']
})

# Compute TF-IDF vectors for both columns
vectorizer = TfidfVectorizer()
tfidf_matrix1 = vectorizer.fit_transform(text_data['text1'])
tfidf_matrix2 = vectorizer.transform(text_data['text2'])

# Compute cosine similarity between corresponding rows
cosine_similarities = []
for i in range(tfidf_matrix1.shape[0]):
    cosine_sim = cosine_similarity(tfidf_matrix1[i], tfidf_matrix2[i])
    cosine_similarities.append(cosine_sim[0][0])

# Add cosine similarity to the DataFrame
data['cosine_similarity'] = cosine_similarities

# Calculate the mean, median, and standard deviation of the cosine similarities
mean_similarity = data['cosine_similarity'].mean()
median_similarity = data['cosine_similarity'].median()
std_similarity = data['cosine_similarity'].std()

# Print the results
print(f"Mean Cosine Similarity: {mean_similarity:.2f}")
print(f"Median Cosine Similarity: {median_similarity:.2f}")
print(f"Standard Deviation of Cosine Similarity: {std_similarity:.2f}")

# Plot the distribution of cosine similarity
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(data['cosine_similarity'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Cosine Similarities')
plt.xlabel('Cosine Similarity')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
